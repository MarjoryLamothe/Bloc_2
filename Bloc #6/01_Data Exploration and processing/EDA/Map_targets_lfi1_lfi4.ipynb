```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_excel("/Users/lamothemarjory/Downloads/Projet Full Stack/big_merge_V2_meteo.xlsx")
```

```python
data_lfi1= data.loc[data['LFI']=='LFI1']
data_lfi4= data.loc[data['LFI']=='LFI4']
```

```python
data_lfi4_lfi1= pd.concat([data_lfi1,data_lfi4], ignore_index=False)
```

```python
data_lfi4_lfi1["SURF_TER_HA_DIFF"]=data_lfi4_lfi1.groupby("PARCELLE")['SURF_TER_HA'].diff()
```

```python
data_lfi4_v2= data_lfi4_lfi1.loc[data_lfi4_lfi1['LFI']=='LFI4']
```

```python
data_lfi4_v2["SURF_TER_HA_DIFF_pos"]=data_lfi4_v2["SURF_TER_HA_DIFF"].apply(lambda x: "Diminution" if x >= 0
else 'Augmentation')
```

```python
data_lfi4_v2["SURF_TER_HA_DIFF_value"] = data_lfi4_v2["SURF_TER_HA_DIFF"].apply(lambda x: abs(x))
fig1 = px.scatter_mapbox(data_lfi4_v2, lat="LAT", lon="LON", size="SURF_TER_HA_DIFF_value", color='SURF_TER_HA_DIFF_pos', zoom=6, mapbox_style="carto-positron")
fig1.show()
```

```python
data_lfi4_v2["ACCR_pos"]=data_lfi4_v2["ACCR"].apply(lambda x: ">0" if x >= 0
else '<0')
```

```python
data_lfi4_v2["ACCR_value"] = data_lfi4_v2["ACCR"].apply(lambda x: abs(x))
fig2 = px.scatter_mapbox(data_lfi4_v2, lat="LAT", lon="LON", size="ACCR_value", color='ACCR_pos', zoom=6, mapbox_style="carto-positron")
fig2.show()
```

```python
data_lfi4_lfi1["TAUX_COUV_RAJ_ASS_pos"] = data_lfi4_lfi1["TAUX_COUV_RAJ_ASS"].apply(lambda x: x if x >= 0
else 0) 
fig3 = px.scatter_mapbox(data_lfi4_lfi1, lat="LAT", lon="LON", size="TAUX_COUV_RAJ_ASS_pos", color='TAUX_COUV_RAJ_ASS_pos', zoom=6, mapbox_style="carto-positron")
fig3.show()
```
